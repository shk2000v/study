# 주사위 게임

1부터 6까지의 눈을 가진 3개의 주사위를 던져서 다음과 같은 규칙에 따라 상금을 받는 게임이 있다.

상금을 출력하는 프로그램

규칙(1) 같은 눈이 3개가 나오면 10,000원+(같은눈)*1,000원의 상금을 박게 된다.

규칙(2) 같은 눈이 2개만 나오는 경우에는 1,000원+(같은 눈) *100원의 상금을 받게 된다.

# 선생님풀이

```python
temp = [6,3,3]
temp.sort()
a, b, c = temp

if a==b and b==c:
    prize = 10000 + (a*1000)
elif a==b or b ==c:
    prize = 1000 + (b*100)
else:
    prize = c*100
print(prize)
```

```python
with open("주사위게임1.txt", "r") as f:
    n =int(f.readline())
    max_prize = 0
    for i in range(n):
        temp =list(map(int,f.readline().split()))
        temp.sort()
        a,b,c = temp
        
        if a==b and b==c:
            prize = 10000 + (a*1000)
        elif a==b or b ==c:
            prize = 1000 + (b*100)
        else:
            prize = c*100
        
        if prize > max_prize:
            max_prize = prize
        
print(max_prize)
        
        
```
