
# 선생님 풀이

```python
from collections import deque
with open("사과나무in1.txt", "r") as f:
   n = int(f.readline())
   matrix = [list(map(int,f.readline().split())) for _ in range(n)]
#초기화 작업

level, total = 0 , 0
dx=[-1,0,1,0]
dy=[0,1,0,-1]
Q = deque()
ck_tbl = [[0]*n for _ in range(n)]
total += matrix[n//2][n//2]
ck_tbl[n//2][n//2] = 1
Q.append((n//2, n//2))

#탈출 조건?


while True:
    if level == n//2:
        break
    size = len(Q)
    for i in range(size):
        temp = Q.popleft()
        for j in range(4):
            x = temp[0]+dx[j] 
            y = temp[1]+dy[j]
            if ck_tbl[x][y] == 0:
                total += matrix[x][y]
                ck_tbl[x][y] = 1
                Q.append((x,y))
    level += 1
print(total)
```

```python
for xx in matrix:
    print(xx)
```

```python
for dd in ck_tbl:
    print(dd)
```
