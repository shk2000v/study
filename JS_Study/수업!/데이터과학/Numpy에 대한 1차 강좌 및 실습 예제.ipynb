Vectorization operation
- 배열화계산
- 원소 하나씩 계산하는 것이 아니라 배열 단위로 계산
- Loops없이 연산하여 코딩의 양이 줄어듬
- 병렬처리가 가능(Multi-core)

- 덧셈뺄셈이 가능하려면 shape(크기)이 같아야 한다
