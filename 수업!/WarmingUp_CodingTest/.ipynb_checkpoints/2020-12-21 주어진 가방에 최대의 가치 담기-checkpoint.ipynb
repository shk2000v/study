```python
def DFS(wsum,vsum):
    global result
    if wsum > limit_kg:
        return
    if vsum >result:
        result =vsum
    for w, v in gem:
        print(wsum, vsum)
        DFS(wsum+w, vsum+v)

f=open("가방담기in1.txt",'r')
n, limit_kg = map(int, f.readline().split())
gem = []
for _ in range(n):
    a,b = map(int, f.readline().split())
    gem.append((a,b))
gem.sort()

result = 0
DFS(0,0)
print(result)

f.close()
```

# DFS 보다 빠른 알고리즘
# 냅색 알고리즘

```python
f=open("가방담기in1.txt",'r')
n, limit_kg = map(int, f.readline().split())
wbag_idx = [0]*(limit_kg+1)
for i in range(n):
    w, v = map(int, f.readline().split())
    for j in range(w, limit_kg+1):
        wbag_idx[j] = max(wbag_idx[j], wbag_idx[j-w]+v) #이거는 꼭 기억하면 된다
        # 인덱스를 w로 받고 그 차만큼
        # 
        print(j, wbag_idx)
print(wbag_idx[limit_kg])
```
