```python
pop_seoul.columns=['구별','인구수','서울시민','외국인','고령자']
```

```python
pop_seoul.set_index('구별',inplace =True)
```

```python
pop_seoul.head()
```

```python
pop_seoul.drop('합계', inplace=True)
```

```python
pop_seoul.head()
```

```python
#첫번째 소수점 처리 방식
pop_seoul['한국인비율(%)'] = np.round((pop_seoul['서울시민']/pop_seoul['인구수'])*100, decimals=2)
# np.round(~~~ , decimas='보여주고싶은 소수n번') 으로 소숫점을 고르는 방법이 있다.


pop_seoul['외국인비율(%)'] = (pop_seoul['외국인']/pop_seoul['인구수'])*100
pop_seoul['고령자비율(%)'] = (pop_seoul['고령자']/pop_seoul['인구수'])*100
```
