# 선생님 풀이

```python
with open("격자판회문수1.txt", "r") as f:
    mat = [list(map(int,f.readline().split())) for _ in range(7)]
count=0
    #가로 탐색
for i in range(3):
    for j in range(7):
        temp = mat[j][i:i+5]
        if temp == temp[::-1] :
            count += 1
        for k in range(2):
            if mat[i+k][j] != mat[i+4-k][j]:
                break
        else : # for else문 이라는게 있다
            count += 1
print(count)
                

    
```
