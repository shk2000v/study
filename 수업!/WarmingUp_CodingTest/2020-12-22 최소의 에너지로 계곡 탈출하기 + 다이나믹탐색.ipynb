```python
import time
start = time.time()

with open("계곡탈출in1.txt", 'r') as f:
    n= int(f.readline())
    ar = [list(map(int,f.readline().split())) for _ in range(n)]
    
dt= [[0]*n for _ in range(n)]    
dt[0][0] = ar[0][0] # dt 의 초기화

for i in range(1,n):
    dt[0][i] = dt[0][i-1] + ar[0][i] #가로세로 더하는걸 업데이트
    dt[i][0] = dt[i-1][0] + ar[i][0]

for i in range(1,n):
    for j in range(1,n):
        dt[i][j] = min(dt[i-1][j], dt[i][j-1])+ ar[i][j] # 둘중에서 min값 하나
print(dt[n-1][n-1])
print(time.time() - start)
```
