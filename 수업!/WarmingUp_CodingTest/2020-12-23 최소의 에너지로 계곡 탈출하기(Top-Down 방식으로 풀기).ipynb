
DFS가 (n-1,n-1)로 시작하는 방법을 구현


```python
with open("계곡탈출in1.txt", 'r') as f:
    n= int(f.readline())
    ar = [list(map(int,f.readline().split())) for _ in range(n)]

    
dt= [[0]*n for _ in range(n)]
dt[n-1][n-1] = ar[n-1][n-1]

for i in range(n-1,0,-1):
    dt[n-1][i-1] = dt[n-1][i-1] + ar[n-1][i-1] #가로세로 더하는걸 업데이트
    dt[i-1][n-1] = dt[i-1][n-1] + ar[i-1][n-1]
    
for i in range(n-1,0,-1):
    for j in range(n-1,0,-1):
        dt[i][j] = min(dt[i-1][j], dt[i][j-1])+ ar[i][j] # 둘중에서 min값 하나
print(dt[0][0])
```

# 선생님 풀이


```python
f=open('계곡탈출in3.txt', 'r')
n= int(f.readline())
ar = [list(map(int,f.readline().split())) for _ in range(n)]

def DFS(row,col):
    if row== 0 and col == 0:
        return ar[row][col]
    else:
        if row==0:
            return DFS(row, col-1) + ar[row][col] 
        elif col ==0:
            return DFS(row-1, col) + ar[row][col]
        else:
            return min(DFS(row-1, col), DFS(row,col-1)) + ar[row][col]

DFS(n-1,n-1)
```
