```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.sex = tips.sex.map({'Female':0, 'Male':1})
```

```python
tips
```

```python
tips.smoker = tips.smoker.map({'No':0,"Yes":1})
```

```python
tips
```

```python
tips.day.unique()
```

```python
tips.day=tips.day.map({'Thur':0, 'Fri':1, 'Sat':2, 'Sun': 3})
```

```python
tips
```

```python
tips.time = tips.time.map({'Dinner':0, 'Lunch':1})
```

```python
tips
```

```python
tips=tips[[	'total_bill','sex',	'smoker',	'day',	'time',	'size','tip']]
```

```python
tips
```

```python
X = tips.iloc[:,0:-1]
```

```python
X
# columns가 6개니깐 weight는 6개
```

```python
y = tips.iloc[:,-1]
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(X,y)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
tips.head(5)
```

```python
LR.predict([[50,0,1,2,1,6]])
```
