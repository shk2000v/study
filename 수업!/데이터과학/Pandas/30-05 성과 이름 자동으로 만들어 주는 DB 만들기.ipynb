```python
names = [ name_gen() for i in range(10)]
weights = [random.randint(45,101) for xx in names]
heights = [random.randint(150,221) for xx in names]
```

```python
BMI= pd.DataFrame({
    "이름" : names,
    "몸무게" : weights,
    "키" : heights
})
```

```python
BMI
```

```python
#BMI_index = 몸무게/(키/100) **2
BMI["BMI_index"] = BMI.몸무게/(BMI.키/100) ** 2
```

```python
BMI
```

```python
BMI.round(decimals=2)
```

```python
def health_info(h_idx):
    if h_idx < 20:
        return '저체중'
    elif h_idx >= 20 and h_idx <= 24:
        return '정상'
    elif h_idx > 24 and h_idx <= 29:
        return '과체중'
    else:
        return '비만'
```

```python
BMI['BMI결과'] = [health_info(xx) for xx in BMI.BMI_index]
# 이런 방식도 있는데 더 권장하는 방법이 있다
# 안되는건 아닌데 
```

```python
BMI.rename(columns={'BMI결과':'BMI결과1'}, inplace=True)
```

```python
BMI['BMI결과2'] = BMI['BMI_index'].apply(health_info)
# mapping과 비슷하여 가장 처리속도가 빠르다
# map 과 lambda는 거의 동급 속력
```
