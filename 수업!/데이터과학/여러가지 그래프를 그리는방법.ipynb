```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
import platform
sns.set(style='darkgrid')

if platform.system() == 'Windows':
# 윈도우인 경우
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:    
# Mac 인 경우
    rc('font', family='AppleGothic')
rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus'] = False
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#그림 스타일 relation 중심 relplot, 분류중심 catplot,  joinplot

sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', row='sex', col='day')
```

```python
sns.relplot(x='total_bill', y='tip',data=tips, hue='time', palette=['b','r'], size='size', col='time')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
df= pd.DataFrame(np.random.randn(500,2), columns=['0','1']) ##정규분포를 땐 상태로 데이터를 랜덤으로
```

```python
df
```

```python
sns.relplot(x='0', y= '1', data=df )
```

```python
sns.jointplot(x='0', y= '1', data=df )
```

```python
sns.jointplot(x='0', y= '1', data=df ,kind='hex')
#벌집모양처럼 보여준다
```

```python
sns.jointplot(x='0', y= '1', data=df ,kind='resid')
# 정규분포선을 보여준다
```

```python
sns.jointplot(x='0', y= '1', data=df ,kind='reg')
```
