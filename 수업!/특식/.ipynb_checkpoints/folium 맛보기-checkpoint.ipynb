```python
df = pd.read_excel("C:/Users/KimSeungHwan/Downloads/korcorona.xlsx")
```

```python
df.index=df.지역
```

```python
df.drop(['지역'], axis=1, inplace=True)
```

```python
df
```

```python
for xx in df.index:
    print(xx)
```

```python
m= folium.Map(location=[37.545998, 126.926246], zoom_start=7, tiles="OpenStreetMap")
mean = df['확진자'].mean()
for xx in df.index:
    lati = df.loc[xx,'위도']
    longi=df.loc[xx,'경도']
    scale=df.loc[xx, '확진자'] // mean
    if df.loc[xx, '확진자'] > mean:
        folium.Circle(location=[lati, longi], radius=(15000scale+10000)
        ,color='#FF0000', fill=True, fill_color='#FF0000').add_to(m)
    else:
        folium.Circle(location=[lati, longi], radius=(15000scale+10000)
        ,color='#FF476600', fill=True, fill_color='#FF476600').add_to(m)

m.save('map.html')

print(df)
```
