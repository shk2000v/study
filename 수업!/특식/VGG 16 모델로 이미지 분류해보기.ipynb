```python
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
import os
import glob  # 윈도우환경을 리눅스처럼 쓸수있게 해주는 명령어
```

```python
model = VGG16()
```

```python
img_path="C:/Media/지옥견세마리.jpg"
img = image.load_img(img_path, target_size=(224,224))
```

```python
img
```

```python
X = image.img_to_array(img)
```

```python
X.shape
```

```python
X=np.expand_dims(X, axis=0)
```

```python
X.shape
```

```python
X = preprocess_input(X)
```

```python
features=model.predict(X)
```

```python
label=decode_predictions(features)
```

```python
print("%s(%.2f%%)"%(label[0][0][1], label[0][0][2]*100))
```

```python
img_path="C:/Media/"
for filename in glob.glob(os.path.join(img_path, '*.*')):
    img = image.load_img(filename, target_size=(224,224))
    X = image.img_to_array(img)
    X=np.expand_dims(X, axis=0)
    X = preprocess_input(X)
    features=model.predict(X)
    label=decode_predictions(features)
    print(filename+"=======>","%s(%.2f%%)"%(label[0][0][1], label[0][0][2]*100))
```
